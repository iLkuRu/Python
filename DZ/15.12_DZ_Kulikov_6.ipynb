## Задача 1. Книги без буквы E.

**Условие.** В истории литературы известен случай написания романа объемом около
50 тыс. слов, в котором ни разу не была употреблена самая популярная
в английском алфавите буква E. Название его – «Gadsby».
Напишите программу, которая будет считывать список слов из файла и собирать статистику о том, в каком проценте слов используется каждая буква алфавита. Выведите результат для всех 26 букв английского алфавита и отдельно отметьте букву, которая встречалась в словах наиболее редко. В  вашей программе должны игнорироваться знаки препинания и регистр символов.


**Подсказки.**
1. Переведите все слова в верхний регистр:
```
word = word.upper().rstrip() # upper - переводит строку в верхний регистр
```
2. Создайте словарь со счетчиком слов, содержащий каждую букву
```
# Для каждой буквы инициализируем счетчик нулем
counts = {}
for ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
       counts[ch] = 0
```
3. Наиболее редко используемое слово можно найти следующим образом:
```
smallest_count = min(counts.values()) # min - возвращает наименьшее число в списке
```

````python
#Напишите свое решение
!wget https://www.gutenberg.org/cache/epub/47342/pg47342.txt

import string
#генерируем словарь букв
counts = {ch: 0 for ch in string.ascii_uppercase}

#читаем файл
file = open('pg47342.txt', 'r', encoding='utf-8')
#создаем список слов в верхнем регистре
read = file.read().upper().split()
#создаем пустой список
words = []
#очистка списка 'read' от ненужных символов, добавление в список words
for word in read:
    b = word.strip('|()-_,.;:!"+')
    words.append(b)

# цикл считает количество каждой буквы в заданных словах и общее количество букв
total_letters = 0
for word in words:
    for char in word:
        if char.isalpha():
            counts[char] += 1
            total_letters += 1

#вычисления процента букв
letter_percentage = {letter: count / total_letters * 100 for letter, count in counts.items()}
rarest_letter = min(letter_percentage, key=letter_percentage.get)

#печать вывода.
print('Статистика использования букв:')
for letter in string.ascii_uppercase:
    print(f'{letter}: {letter_percentage[letter]:.2f}%')
print(f'\nСамая редкая буква: {rarest_letter}')
````

````python
import string

# генерируем словарь букв
!wget https://www.gutenberg.org/cache/epub/47342/pg47342.txt

counts = {ch: 0 for ch in string.ascii_uppercase}
#читаем файл
file = open('pg47342.txt', 'r', encoding='utf-8')
#создаем список слов в верхнем регистре
read = file.read().upper().split()
# список слов в верхнем регистре
words = [word.strip('|()-_,.;:!"+') for word in read]

# общее количество слов
total_words = len(words)

# цикл считает количество слов, в которых встречается каждая буква
for word in words:
    for char in set(word):  # используем set, чтобы каждая буква в слове учитывалась только один раз
        if char.isalpha():
            counts[char] += 1

# вычисления процента слов, в которых встречается каждая буква
letter_percentage = {letter: count / total_words * 100 for letter, count in counts.items()}
rarest_letter = min(letter_percentage, key=letter_percentage.get)

# печать вывода.
print('Статистика использования букв:')
for letter in string.ascii_uppercase:
    print(f'{letter}: {letter_percentage[letter]:.2f}%')
print(f'\nСамая редкая буква: {rarest_letter}')
````

## Задание 2. Веб-скрейпинг ИИ новостей

**Описание задания.** Используя пример веб-скрейпинга практической части урока, проведите скрейпинг сайта с новостями из сферы ИИ (https://2051.vision/category/ii/), выведите на экран заголовки новостей.

````python
#Напишите свое решение
from bs4 import BeautifulSoup
import requests

url = ("https://2051.vision/category/ii/") # Зададим адрес новостного сайта для GET-запроса библиотеки requests
html = requests.get(url).text # Извлекаем из тела ответа текстовые данные
soup = BeautifulSoup(html, 'html5lib') # Применяем к данным анализатор html5lib
list =[]

top_news = soup.find_all('a', class_="td-image-wrap")

for data in top_news:
    if 'title' in data.attrs:
        title = data.get('title')
        list.append(title)

for i, news in enumerate(list, 1):  # Используем enumerate с параметром start=1, чтобы индексация начиналась с 1
    print(f"{i}. {news}")
````
