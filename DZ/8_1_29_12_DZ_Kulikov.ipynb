Ответь на вопрос: игры каких жанров были наиболее популярны до 2000 года, а какие после?:

```python
df_before_2000 = df[df['Year'] <= 2000]
df_after_2000 = df[df['Year'] > 2000]

sns.set(style="darkgrid", font_scale=1.0)
plt.figure(figsize=(10, 5))

ax1 = sns.barplot(x='Global_Sales', y='Genre', data=df_before_2000, estimator=sum, errorbar=None, palette='viridis', order=df_before_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).index)
ax1.set_title('Популярность жанров игр до 2000 года', fontsize=16)
ax1.set_xlabel('Глобальные продажи (млн)', fontsize=14)
ax1.set_ylabel('Жанр', fontsize=14)
plt.show()

print()

plt.figure(figsize=(10, 5))
ax2 = sns.barplot(x='Global_Sales', y='Genre', data=df_after_2000, estimator=sum, errorbar=None, palette='viridis', order=df_after_2000.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False).index)
ax2.set_title('Популярность жанров игр после 2000 года', fontsize=16)
ax2.set_xlabel('Глобальные продажи (млн)', fontsize=14)
ax2.set_ylabel('Жанр', fontsize=14)
plt.show()
```

```python
df['Year'] = df['Year'].fillna(0).astype(int)
```

```python
df = df.query('Year != 0').reset_index(drop=True)
```

```python

sales_per_year = df.groupby('Year')['Global_Sales'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Global_Sales', data=sales_per_year)
plt.xlabel('Год выпуска')
plt.ylabel('Общее число продаж (млн)')
plt.title('Общее число видеоигр по годам')
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

```python
top_publishers = df['Publisher'].value_counts().nlargest(3).index.tolist()

df_top_publishers = df[df['Publisher'].isin(top_publishers)]

fig = px.bar(df_top_publishers, x='Platform', color='Publisher',
             title='Количество выпущенных игр для каждой платформы (топ-3 издателей)',
             labels={'Name': 'Количество игр', 'Platform': 'Платформа'},
             barmode='stack',
             color_discrete_map={'Publisher1': 'lightcoral', 'Publisher2': 'lightblue', 'Publisher3': 'lightgreen'})

fig.update_layout(
    xaxis=dict(title_text='Платформа', tickmode='linear'),
    yaxis=dict(title_text='Количество игр'),
    legend_title_text='Издатель:',
    showlegend=True
)

fig.show()
```

```python
df_1980_2000 = df[(df['Year'] >= 1980) & (df['Year'] <= 2000)]
df_2000_2020 = df[(df['Year'] > 2000) & (df['Year'] <= 2020)]

sales_columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
na_eu_jp_sales_1980_2000 = df_1980_2000[sales_columns].sum()
na_eu_jp_sales_2000_2020 = df_2000_2020[sales_columns].sum()

total_global_sales_1980_2000 = na_eu_jp_sales_1980_2000['Global_Sales']
region_sales_share_1980_2000 = na_eu_jp_sales_1980_2000 / total_global_sales_1980_2000

total_global_sales_2000_2020 = na_eu_jp_sales_2000_2020['Global_Sales']
region_sales_share_2000_2020 = na_eu_jp_sales_2000_2020 / total_global_sales_2000_2020

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

axes[0, 0].pie(region_sales_share_1980_2000[:-1], labels=region_sales_share_1980_2000.index[:-1], autopct='%1.1f%%')
axes[0, 0].set_title('1980-2000 гг.')

axes[0, 1].pie(region_sales_share_2000_2020[:-1], labels=region_sales_share_2000_2020.index[:-1], autopct='%1.1f%%')
axes[0, 1].set_title('2000-2020 гг.')

axes[1, 0].axis('off')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df1_sorted, x='Nobel Prices', y='Gross National Income', hue='Average IQ', palette='viridis', s=100)

plt.title('Нобелевская премия относительно валового национального дохода со средним IQ')
plt.xlabel('Нобелевская премия')
plt.ylabel('валовый национальный доход')

plt.legend(bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```

```python
df1_sorted['Mean years of schooling'] = df1_sorted['Mean years of schooling'].fillna(df1_sorted['Mean years of schooling'].mean())

fig = px.scatter(df1_sorted, x='Nobel Prices', y='Gross National Income', color='Average IQ',
                 size='Mean years of schooling', hover_data=['Country'],
                 labels={'Country': 'Страна',
                         'Gross National Income': 'Валовый национальный доход',
                         'Average IQ': 'Средний IQ',
                         'Mean years of schooling': 'Средний срок обучения',
                         'Nobel Prices': 'Нобелевская премия'},
                 title='Нобелевская премия относительно валового национального дохода со средним IQ и средним сроком обучения.',
                 size_max=10)
fig.show()
```
